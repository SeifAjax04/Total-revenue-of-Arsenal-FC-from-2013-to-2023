```python
import matplotlib.pyplot as plt
```

```python
# Data
years = ["2013-14", "2014-15", "2015-16", "2016-17", "2017-18", "2018-19", "2019-20", "2020-21", "2021-22", "2022-23"]
revenue = [359, 436, 469, 488, 439, 445, 388, 367, 433, 533]  # in millions (€)
```

```python
# Find the index of the minimum and maximum revenue
min_revenue_index = revenue.index(min(revenue))
max_revenue_index = revenue.index(max(revenue))
```

```python
# Create the plot
plt.figure(figsize=(12, 8))  # Adjust the figure size as needed
plt.plot(years, revenue, marker='o')

plt.title('Total revenue of Arsenal FC from 2013/14 to 2022/23', color='red', weight='bold')
plt.xlabel('Year')
plt.ylabel('Revenue (Millions €)')
plt.xticks(rotation=45)
plt.grid(True)

# Remove the grid
plt.grid(False)

# Add text labels in boxes for each revenue point
for i, rev in enumerate(revenue):
    plt.text(years[i], rev, f"{rev}M", color='black', ha='center', va='bottom', bbox=dict(boxstyle="square,pad=0.3", fc="white", ec="black", lw=1))

# Add annotation with arrow pointing to the minimum revenue
plt.annotate(f"Lowest Revenue: €{min(revenue)}M ({years[min_revenue_index]})", 
             xy=(min_revenue_index, min(revenue)), xycoords='data',  # Position of the arrow
             xytext=(-70, -18), textcoords='offset points',  # Position of the text
             arrowprops=dict(facecolor='red', arrowstyle="->", color='red'),  # Arrow color
             bbox=dict(boxstyle="square,pad=0.3", fc="white", ec="black", lw=1))  # Add a box around the text

# Add annotation with arrow pointing to the maximum revenue
plt.annotate(f"Highest Revenue: €{max(revenue)}M ({years[max_revenue_index]})", 
             xy=(max_revenue_index, max(revenue)), xycoords='data',  # Position of the arrow
             xytext=(-83, 40), textcoords='offset points',  # Position of the text (adjusted to move left)
             arrowprops=dict(facecolor='red', arrowstyle="->", color='red'),  # Arrow color
             bbox=dict(boxstyle="square,pad=0.3", fc="white", ec="black", lw=1))  # Add a box around the text

# Save the plot as a PNG image
plt.savefig('Total revenue of Arsenal FC.png', dpi=300)

# Display the plot
plt.show()
```
